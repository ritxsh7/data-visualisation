```python
import seaborn as sns

flights = sns.load_dataset('flights')
flights.head()
```

```python
#matrix plotting requires variables in 2d matrix

flights_mat = flights.pivot(index = 'month', columns=['year'], values='passengers')
flights_mat
```

```python
sns.heatmap(
    data=flights_mat, 
    annot=False, #shows/hides actual value 
    # cmap='coolwarm' #pallete
    )
```

```python
sns.clustermap(data=flights_mat)
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
#Grids

# 1. PairGrid : grid plots for every pair in the dataset

g = sns.PairGrid(iris)

g.map(sns.scatterplot)
```

```python
#FacetGrid plot: separates a plot according to other variables

tips = sns.load_dataset('tips')

g = sns.FacetGrid(
    data=tips,
    col='time',
    row = 'smoker'
)

g.map(sns.barplot, 'total_bill')
```

```python
#linear regression plots
tips.head()
```

```python
sns.lmplot(
    tips, 
    x = 'total_bill', 
    y = 'tip',
    # hue='sex', #highlight categories
    col='sex', #separate by category
    )
```
