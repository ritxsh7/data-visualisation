```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
#Functional method

plt.plot(x,y)
plt.show()
```

```python
#Object oriented method

fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8, 0.8])
axes1.plot(x,y)
```

```python
#subplots 

fig, axes = plt.subplots(nrows = 1, ncols = 2)
```

```python
for axis in axes:
    axis.plot(x,y)

fig
```

```python
#custom layouts

fig = plt.figure(figsize=(8,4))
fig
```
