```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#CATEGORICAL DATA

# 1. bar plot: groups the data by category(x axis) and an estimator value on y axis(default = avg)

# sns.barplot(x='sex', y='total_bill', data=tips)
sns.barplot(x = 'day', y = 'total_bill', data=tips, estimator=np.std)
```

```python
# 2. countplot: similar to barplot but the estimator is count by default

sns.countplot(x='smoker', data=tips)
```

```python
# 3. boxplot: shows the data in 4 quartiles and outliers

sns.boxplot(x = 'sex', y = 'tip', data=tips)
```

```python
#options for boxplot

sns.boxplot(x = 'sex', y = 'tip', data=tips, hue='smoker')
```

```python
# 4. striplot: similar to bar plot but shows the entire data instead of range and points may overlap

sns.stripplot(
    x='day', 
    y = 'tip', 
    data=tips, 
    jitter=True #scatters points for visibilty
)
```
