```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#distribution using seaborn

# 1. distribution plot (histogram)

sns.distplot(tips["total_bill"], bins=30, kde=False)
plt.show()
```

```python
# 2. jointplot: used for comparing 2 variables of the dataset
# lets compare total bill and tip variables

sns.jointplot(x='total_bill', y = 'tip', data=tips)
plt.show()
```

```python
#kind param allows kind of plot to draw like hex, reg, kde, etc

sns.jointplot(x='total_bill', y = 'tip', data=tips, kind='reg')
plt.show()
```

```python
# pairplot: allows to plot between each of the numerical pairs in the dataset

sns.pairplot(tips)
plt.show()
```

```python
#options in pairplot

#hue: for categorical data
#pallete: for color scheme
sns.pairplot(tips, hue='sex', palette="coolwarm")
plt.show()
```

```python
# 4. rugplot: distribution plot but shows a bar on datapoint each time it appears

sns.rugplot(tips['tip'])
plt.show()
```
